```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.patches as patches
from matplotlib.gridspec import GridSpec, GridSpecFromSubplotSpec
```

```python
sns.reset_defaults()

def graphs():
    
    #creating random numpy array 100x2 between 10-100
    new_array = np.random.randint(10,100, size=(100,2))
    
    #2x2 figure creating
    fig, ax = plt.subplots(2, 2, figsize=[13,10])

    #creating line plot and set labels of x-y axis and title. x axis limits starting from -5 value to 105. 
    #Ticks sets 0-100 with 10 stepsize
    ax[0,0].set_title("Line Plot")
    ax[0,0].set_xlabel("Random var 1")
    ax[0,0].set_ylabel("Value")
    ax[0,0].set_xticks(np.arange(0, 101,10))
    ax[0,0].set_xlim([-5, 105])
    
    #creating line plot sorted depending on first values and then plot
    ax[0,0].plot(new_array[new_array[:, 0].argsort()][:,0], new_array[new_array[:, 0].argsort()][:,1])

    #creating scatter plot and set labels of x-y axis and title. x axis limits starting from 0 value to 100. 
    #Grid line style also set.
    ax[0,1].set_title("Scatter Plot")
    ax[0,1].set_xlabel("Random var 1")
    ax[0,1].set_ylabel("Random var 2")
    ax[0,1].grid(linestyle = "-")
    ax[0,1].set_xlim([0, 100])
    ax[0,1].set_ylim([0, 100])
    
    #creating scatter plot without sorting
    ax[0,1].scatter(new_array[:,0], new_array[:,1])

    #creating histogram and set labels of x-y axis and title. x axis limits starting from 0 value to 100 with stepsize 20.
    ax[1,0].set_title("Histogram Plot")
    ax[1,0].set_xlabel("Random var 1")
    ax[1,0].set_ylabel("Frequency")
    ax[1,0].hist(new_array[:,0])

    #creating bar chart and set labels of x-y axis and title. x ticks set 1-10 and y axis 0-100 with stepsize 20
    ax[1,1].set_title("Bar Chart")
    ax[1,1].set_xlabel("Random var 1")
    ax[1,1].set_ylabel("Value")
    ax[1,1].set_xticks(np.arange(1, 11))
    ax[1,1].set_yticks(np.arange(0, 101, 20))
    ax[1,1].set_ylim([0, 100])
    #took first 10 values of array
    ax[1,1].bar(np.arange(1,11), new_array[:,1][:10])
    
    return plt.gca()
```

```python
#call function
graph = graphs()

# displays the plot
plt.show()
```
