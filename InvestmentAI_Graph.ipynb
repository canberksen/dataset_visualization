```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
def investment_graph(data):

    fig,ax = plt.subplots()

    #x variable for Year values, y variable for inflation values
    x=data["Year"]
    y=data["total_corporate_investment_inflation_adjusted"]

    #mark values with *
    ax.plot(x, y, marker="*")

    # scale values for billion
    scale_y = 1e9
    ticks_y = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/scale_y)+" billion$")
    ax.yaxis.set_major_formatter(ticks_y)

    # arrange y axis values 0 to 180 billion by step size 20 billion 
    plt.yticks(np.arange(0, 180000000001, 20000000000))

    # add gridlines, grid line style. Title of graph and make left,
    # right, top spines unvisible
    plt.grid(axis="y",linestyle = "--")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    
    return plt.gca()
```

```python
# import data from csv file
data = pd.read_csv(r"corporate-investment-in-artificial-intelligence-total.csv")

#call function with imported data
graph = investment_graph(data)

# displays the plot
plt.show()
```
